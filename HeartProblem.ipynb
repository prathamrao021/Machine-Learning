```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("/content/heart_2020_cleaned.csv")
data.shape
```

```python
data = data.drop(columns = ["Race"])
data
```

```python
data.isnull().sum()
```

```python
data['Smoking'] = data['Smoking'].map({'Yes':1,'No':0})
data
```

```python
data['HeartDisease'] = data['HeartDisease'].map({'Yes':1,'No':0})
data['AlcoholDrinking'] = data['AlcoholDrinking'].map({'Yes':1,'No':0})
data['Stroke'] = data['Stroke'].map({'Yes':1,'No':0})
data['DiffWalking'] = data['DiffWalking'].map({'Yes':1,'No':0})
data['Diabetic'] = data['Diabetic'].map({'Yes':1,'No':0})
data['PhysicalActivity'] = data['PhysicalActivity'].map({'Yes':1,'No':0})
data['Asthma'] = data['Asthma'].map({'Yes':1,'No':0})
data['KidneyDisease'] = data['KidneyDisease'].map({'Yes':1,'No':0})
data['SkinCancer'] = data['SkinCancer'].map({'Yes':1,'No':0})
data
```

```python
data['GenHealth'].value_counts()
```

```python
data['GenHealth'] = data['GenHealth'].map({'Excellent':5,'Very good':4,'Good':3,'Fair':2,'Poor':1})
data['GenHealth'].describe()
```

```python
data['GenHealth'].value_counts()
```

```python
data.head()
```

```python
data['Sex'] = data['Sex'].map({'Male':1,'Female':0})
data.head()
```

```python
data['AgeCategory'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
x = data['AgeCategory'].values
data['AgeCategory'] = lb.fit_transform(x)
x = x.reshape(-1,1)
```

```python
data.head()
```

```python
data.corr()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data[['BMI','Smoking','Stroke','PhysicalHealth','MentalHealth','DiffWalking','Sex','AgeCategory','Asthma','KidneyDisease','SkinCancer']]
y = data[['HeartDisease']]

x_train,x_test,y_train,y_test = train_test_split(X, y, test_size=0.25)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
data.describe()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 23)
knn.fit(x_train,y_train)
predictions = knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
accuracy = accuracy_score(y_test,predictions)
print(accuracy)

target_names = ['HeartProblem','NotAHeartProblem']
print(classification_report(y_test, predictions.reshape(-1,1), target_names=target_names))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay,confusion_matrix
import matplotlib.pyplot as plt
cm = confusion_matrix(y_test, predictions.reshape(-1,1), normalize='true')
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels=target_names)
fig, ax = plt.subplots(figsize = (8,6))
disp.plot(ax=ax, cmap="Blues")
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(x_train,y_train)

predictions = svc.predict(x_test)
print(predictions)
```
